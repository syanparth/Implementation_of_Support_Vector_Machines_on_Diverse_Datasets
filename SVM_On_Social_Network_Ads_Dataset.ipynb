***SVM on Social Network Ads dataset***

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
x=df.iloc[:,[2,3]]
y=df.iloc[:,[4]]
```

```python
x.head() # it is independent variable
```

```python
y.head() # it is dependent variable
```

```python
# Splitting data into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
 x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='linear')                  # Linear Kernel
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
```

```python
y_pred # it contains 80 test  values
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred) # here we check the difference between actual "y-values" in "y-test" and  predicted values in [y pred]
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf',random_state=0 ,gamma=7 , C=7)   # low  value of "gama" means we are calculating distance to hyperplane from almost every datapoint.
classifier.fit(x_train,y_train)                                 # high value of "gama" means we are calculating distance to hyperplane from 'closest' datapoint only.
y_pred = classifier.predict(x_test)                             ## C is the regularization parameter which means "adding some value to error function to improve the results or to tell svm how much miss-classification to avoid" 'higher the value of C more accurate the results are'
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)                                    # Rbf Kernel
```

```python
svc=SVC(kernel='poly', degree=3 ) # for non linear data
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)                                         # Polynomial Kernel
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
